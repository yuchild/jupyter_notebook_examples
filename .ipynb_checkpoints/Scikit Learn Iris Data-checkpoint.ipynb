# K-nearest neighbors (KNN) classification
1. Pick a value for K.
2. Search for the K observations in the training data thare are "nearest" to the measurements of the unknown iris
3. Use the most popular response value from the K nearest neighbors as the predicated response value for the unknown iris.

# scikit-learn 4-step modeling pattern
**Step 1**: Import the class you plan to use 

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Step 2**: Instantiate the estimator
- "estimator" is scikit-learn's term for model
- "instantiate" means make an instance of

```python
knn = KNeighborsClassifier(n_neighbors=1) # it is now an object that knows how to KNN
```

- Name of the object does not matter
- Can specify tuning parameters (aka 'hyperparameters') during this step
- Al arameters note specified are set to their defaults

```python
print(knn)
```

**Step 3**: Fit the model with data (aka 'Model Training')
- Model is learning the replationship between x and y
- Occurs in-place

```python
knn.fit(x,y)
```

**Step 4**: Predict the response fro a new observation
- New observations are called "out-of-sample" data
- Uses the information it learned during the model training process

```python
knn.predict(np.array([[3, 5, 4, 2]])) # 2 means virginica predicted
```

- Returns a NumPy array
- Can predict for multiple observations at once

```python
knn.predict(np.array([[3, 5, 4, 2]
                      , [5, 4, 3, 2]])) # virginica, versicolor predicted
```

# Using a different value for K

```python
knn.fit(x,y)
```

```python
knn.predict(np.array([[3,5,4,2]
                     , [5,4,3,2]])) # returns both versicolors
```
